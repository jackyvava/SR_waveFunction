```python
from scipy.io import loadmat
import numpy as np
import numpy as np
from scipy.io import savemat
```

```python
data = loadmat('D:\\zjPhD\\Programzj\\psiToU\\Buquan\\vortex_Tube\\data.mat')
```

```python
# 打印出所有在 .mat 文件中保存的变量的名称
print("Variables stored in data.mat:")
for var_name in data.keys():
    # 排除 loadmat 返回的内置 MATLAB 元数据
    if not var_name.startswith('__'):
        print(var_name)
```

```python
u1 = data['u1']
u2 = data['u2']
u3 = data['u3']

w1 = data['w1']
w2 = data['w2']
w3 = data['w3']

print(u1.shape), print(u2.shape), print(u3.shape)
print(w1.shape), print(w2.shape), print(w3.shape)
```

```python
w = np.sqrt(w1**2 + w2**2 + w3**2)
```
