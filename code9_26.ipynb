```python
import numpy as np
import matplotlib.pyplot as plt

# 创建一个包含三个子图的组图
fig, axs = plt.subplots(1, 3, figsize=(15, 5))

# 加载并显示输入图像
input_data = np.load("D:/zjPhD/Programzj/Diffusion-based-Fluid-Super-resolution/experiments/kmflow_re1000_rs256_ddim_new/recons_u3232_t240_r30_lam0.0/sample_batch63/input_arr.npy")
axs[0].imshow(input_data[0], cmap='viridis')
axs[0].set_title('Input Image')
axs[0].axis('off')
fig.colorbar(axs[0].imshow(input_data[0], cmap='viridis'), ax=axs[0])

# 加载并显示重建图像
reconstructed_data = np.load("D:/zjPhD/Programzj/Diffusion-based-Fluid-Super-resolution/experiments/kmflow_re1000_rs256_ddim_new/recons_u3232_t240_r30_lam0.0/sample_batch63/sample_arr_run_0_it0.npy")
axs[1].imshow(reconstructed_data[0], cmap='viridis')
axs[1].set_title('Reconstructed Image')
axs[1].axis('off')
fig.colorbar(axs[1].imshow(reconstructed_data[0], cmap='viridis'), ax=axs[1])

# 加载并显示参考图像
reference_data = np.load("D:/zjPhD/Programzj/Diffusion-based-Fluid-Super-resolution/experiments/kmflow_re1000_rs256_ddim_new/recons_u3232_t240_r30_lam0.0/sample_batch63/reference_arr.npy")
axs[2].imshow(reference_data[0], cmap='viridis')
axs[2].set_title('Reference Image')
axs[2].axis('off')
fig.colorbar(axs[2].imshow(reference_data[0], cmap='viridis'), ax=axs[2])

# 显示组图
plt.tight_layout()
plt.savefig("compare.png", dpi=300)
plt.show()
```
