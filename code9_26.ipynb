```python
import numpy as np
import matplotlib.pyplot as plt

# 
num_particles = 3
num_points = 100

# 
t = np.linspace(0, 10, num_points)

# 
trajectories = []

for i in range(num_particles):
    # 
    x = np.cumsum(np.random.randn(num_points))
    y = np.cumsum(np.random.randn(num_points))
    z = np.cumsum(np.random.randn(num_points))
    trajectories.append((x, y, z))

# 
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

# 
for i, (x, y, z) in enumerate(trajectories):
    ax.plot(x, y, z, label=f'Particle {i+1}')

# 
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')
ax.set_title('3D Particle Trajectories')

# 
ax.legend()

# 
plt.show()
```
